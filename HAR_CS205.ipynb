```python
def jerk(acc, ts):
    jk = [0]* len(acc)
    for i in range(1,len(acc)):
        jk[i-1] = 1000*(acc[i] - acc[i-1])/(ts[i] - ts[i-1])
    return jk
    
```

```python
def magnitude(num):
    return math.sqrt(num.real**2 + num.imag**2)
```
